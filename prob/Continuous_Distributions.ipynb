```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform, expon, norm, laplace, gamma, beta
```

```python
xvals = np.linspace(-2, 5, 1000)
yvals = uniform.pdf(xvals, 1, 2)
# pdf(x, loc=0, scale=1)
# what do "loc" and "scale" mean?
# U(loc, loc + scale)
plt.plot(xvals, yvals)
plt.title('Uniform(1, 3)')
plt.ylabel('f(x)')
plt.xlabel('x');
```

```python
xvals = np.linspace(0, 50, 5000)
yvals = expon.pdf(xvals, 0, 1)
plt.plot(xvals, yvals)
# pdf(x, loc=0, scale=1)
# what do "loc" and "scale" mean?
# f(x) = (1/scale) * exp(-(x - loc) / scale)
# scale = 1/λ
plt.title('Exp(1)')
plt.ylabel('f(x)')
plt.xlabel('x');
```

```python
xvals = np.linspace(0, 50, 5000)
yvals = expon.pdf(xvals, 0, 10)
plt.plot(xvals, yvals)
# pdf(x, loc=0, scale=1)
# what do "loc" and "scale" mean?
# f(x) = (1/scale) * exp(-(x - loc) / scale)
# scale = 1/λ
plt.title('Exp(1/10)')
plt.ylabel('f(x)')
plt.xlabel('x');
```

```python
xvals = np.linspace(-20, 50, 5000)
yvals = norm.pdf(xvals, 25, 10)
plt.plot(xvals, yvals)
# pdf(x, loc=0, scale=1)
# what do "loc" and "scale" mean?
# N(loc, scale)
# loc = μ, scale = σ
plt.title('N(25, 10^2)')
plt.ylabel('f(x)')
plt.xlabel('x');
```

```python
xvals = np.linspace(-20, 50, 5000)
yvals = norm.pdf(xvals, 10, 5)
plt.plot(xvals, yvals)
# pdf(x, loc=0, scale=1)
# what do "loc" and "scale" mean?
# N(loc, scale)
# loc = μ, scale = σ
plt.title('N(10, 5^2)')
plt.ylabel('f(x)')
plt.xlabel('x');
```

```python
xvals = np.linspace(-20, 50, 5000)
yvals = laplace.pdf(xvals, 10, 5)
plt.plot(xvals, yvals)
# pdf(x, loc=0, scale=1)
# what do "loc" and "scale" mean?
# Laplace(loc, scale)
# loc = a, scale = b
plt.title('Laplace(10, 5)')
plt.ylabel('f(x)')
plt.xlabel('x');
```

```python
xvals = np.linspace(0, 50, 5000)
yvals = gamma.pdf(xvals, 10, 0, 1)
plt.plot(xvals, yvals)
# pdf(x, a, loc=0, scale=1)
# note the extra 'a' parameter!
# what do "loc" and "scale" mean?
# Gamma(x, a, loc, scale) = Gamma((x - loc) / scale, a) / scale
# Where Gamma(a, loc=0, scale=1) = x^(a-1) * e^(-x) / Γ(a)
# scale = 1/β
# a = α
plt.title('Gamma(10, 1)')
plt.ylabel('f(x)')
plt.xlabel('x');
```

```python
xvals = np.linspace(0, 1, 100)
a = 3
b = 10
yvals = beta.pdf(xvals, a, b)
plt.plot(xvals, yvals)
# pdf(x, a, b, loc=0, scale=1)
# note the extra 'a' and 'b' parameters!
# what do "loc" and "scale" mean? let's just ignore...
plt.title('Beta(3, 10)')
plt.ylabel('f(x)')
plt.xlabel('x');
```
