```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli, binom, geom, poisson
```

```python
xvals = np.arange(-5, 5)
yvals = bernoulli.pmf(xvals, 0.2)
plt.bar(xvals, yvals)
plt.title('Bernoulli(0.2)')
plt.ylabel('P(X=x)')
plt.xlabel('x');
```

```python
xvals = np.arange(-5, 15)
yvals = binom.pmf(xvals, n=10, p=0.5)
plt.bar(xvals, yvals)
plt.title('Binomial(10, 0.5)')
plt.ylabel('P(X=x)')
plt.xlabel('x');
```

```python
xvals = np.arange(-10, 110)
yvals = binom.pmf(xvals, n=100, p=0.2)
plt.bar(xvals, yvals)
plt.title('Binomial(100, 0.2)')
plt.ylabel('P(X=x)')
plt.xlabel('x');
```

```python
xvals = np.arange(100)
yvals = geom.pmf(xvals, p=0.1)
# note! scipy uses the k=total # of trials version (vs. total failures)
# So P(X=0) = 0
plt.bar(xvals, yvals)
plt.title('Geometric(0.1)')
plt.ylabel('P(X=x)')
plt.xlabel('x');
```

```python
xvals = np.arange(100)
yvals = geom.pmf(xvals, p=0.9)
# note! scipy uses the k=total # of trials version (vs. total failures)
# So P(X=0) = 0
plt.bar(xvals, yvals)
plt.title('Geometric(0.9)')
plt.ylabel('P(X=x)')
plt.xlabel('x');
```

```python
xvals = np.arange(10)
yvals = poisson.pmf(xvals, mu=0.1)
plt.bar(xvals, yvals)
plt.title('Poisson(0.1)')
plt.ylabel('P(X=x)')
plt.xlabel('x');
```
