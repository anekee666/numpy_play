```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform, expon, norm, laplace, binom, geom, \
  multivariate_normal as mvn
```

```python
xvals = np.linspace(-2, 5, 1000)
yvals = uniform.cdf(xvals, 1, 2)
# pdf(x, loc=0, scale=1)
# what do "loc" and "scale" mean?
# U(loc, loc + scale)
plt.plot(xvals, yvals)
plt.title('Uniform(1, 3)')
plt.ylabel('F(x)')
plt.xlabel('x');
```

```python
xvals = np.linspace(-5, 50, 5000)
yvals = expon.cdf(xvals, 0, 10)
plt.plot(xvals, yvals)
# pdf(x, loc=0, scale=1)
# what do "loc" and "scale" mean?
# f(x) = (1/scale) * exp(-(x - loc) / scale)
# scale = 1/λ
plt.title('Exp(1/10)')
plt.ylabel('F(x)')
plt.xlabel('x');
```

```python
xvals = np.linspace(-20, 50, 5000)
yvals = norm.cdf(xvals, 10, 5)
plt.plot(xvals, yvals)
# pdf(x, loc=0, scale=1)
# what do "loc" and "scale" mean?
# N(loc, scale)
# loc = μ, scale = σ
plt.title('N(10, 5^2)')
plt.ylabel('F(x)')
plt.xlabel('x');
```

```python
xvals = np.linspace(-20, 50, 5000)
yvals = laplace.cdf(xvals, 10, 5)
plt.plot(xvals, yvals)
# pdf(x, loc=0, scale=1)
# what do "loc" and "scale" mean?
# Laplace(loc, scale)
# loc = a, scale = b
plt.title('Laplace(10, 5)')
plt.ylabel('F(x)')
plt.xlabel('x');
```

```python
xvals = np.arange(100)
yvals = binom.cdf(xvals, n=100, p=0.2)
plt.plot(xvals, yvals)
plt.title('Binomial(100, 0.2)')
plt.ylabel('P(X<=x)')
plt.xlabel('x');
```

```python
xvals = np.arange(-5, 10)
yvals = binom.cdf(xvals, n=5, p=0.5)
plt.plot(xvals, yvals)
plt.title('Binomial(5, 0.5)')
plt.ylabel('P(X<=x)')
plt.xlabel('x');
```

```python
xvals = np.arange(100)
yvals = geom.cdf(xvals, p=0.1)
# note! scipy uses the k=total # of trials version (vs. total failures)
# So P(X=0) = 0
plt.plot(xvals, yvals)
plt.title('Geometric(0.1)')
plt.ylabel('P(X<=x)')
plt.xlabel('x');
```

```python
m = [0, 0]
C = [[1, 0.5], [0.5, 1]]

# instead of plotting the PDF (a 3-D plot in this case)
# let's make a 2-D scatterplot of SAMPLES to see how the distributon behaves
xvals = mvn.rvs(m, C, size=10000)

# otherwise matplotlib will make horizontal axis bigger
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.gca().set_aspect('equal')

plt.scatter(xvals[:, 0], xvals[:, 1], alpha=0.2);
```
